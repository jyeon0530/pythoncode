```python
def apply_func_sample(data):
    print(type(data))
    print(data)
    print('-'*10)
```

```python
ktx_df['경전선 KTX'].where(lambda x:x>=5000,'x<5000')
ktx_df['경부선 KTX'].where(lambda x:x<5000,'>=5000')
```

```python
def categorizing_by_5000(data):
    data = data.where(lambda x:x>=5000,0)#where절 true일경우 그대로 false일때 바꿔줌
    data = data.where(lambda x:x<5000,1)
    return data
                      
```

```python
ktx_df.iloc[:,:-2].apply(categorizing_by_5000)
```

```python
ktx_df.apply(apply_func_sample)#벡터 단위로 반복
```
